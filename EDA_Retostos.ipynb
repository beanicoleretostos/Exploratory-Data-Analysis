```python
# Import all the libraries needed for the program to work

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### **Dataset Overview**

```python
# Load the csv file needed for the program

data = pd.read_csv('spotify-2023SONGS.csv', encoding= 'ISO-8859-1')
data
```

### **Determining Datatypes of each column**

```python
print(data.dtypes)
```

### **Checking to see null values**

```python
print(data.isnull().values.any()) #This will return true if there is/are missing/null value.
```

```python
print(data.isnull().sum().sum())
```

```python
pd.isnull(data).sum()
```

```python
data.info()
```

### **Cleaning the dataset**

```python
data['streams'] = pd.to_numeric(data['streams'], errors='coerce')
```

```python
clean_data = data.dropna(subset=['streams'])
clean_data
```

### **Getting the Mean, Median, Standard Deviation of the streams**

```python
print("Mean:", clean_data['streams'].mean())
```

```python
print("Median:", clean_data['streams'].median())
```

```python
print("SD:", clean_data['streams'].std())
```

### **Summary of statistics**

```python
summary = data.describe()
summary
```

### **Top 5 Streamed**

```python
top_5_streamed = data.sort_values('streams', ascending=False).head(5)
print(top_5_streamed)
```

### **Top 5 Artists**

```python
top_5_a = data['artist(s)_name'].value_counts().head(5)
print(top_5_a)
```

### **Tracks Released per Year (Graph)**

```python
# Plot the value counts directly as a bar chart
data['released_year'].value_counts().sort_index().plot.bar()

# Customize the graph labels and title
plt.title('Tracks Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')

# Display the plot
plt.show()
```

### **Tracks Released per Month (Graph)**

```python
# Plot the value counts directly as a bar chart
data['released_month'].value_counts().sort_index().plot.bar()

# Customize the graph labels and title
plt.title('Tracks Released per Month')
plt.xlabel('Month')
plt.ylabel('Number of Tracks')

# Display the plot
plt.show()
```

### **Correlation of the Attributes people are looking for**

```python

correlation = data.filter(items=['streams', 'bpm', 'danceability_%', 'energy_%']).corr()

# Print the correlation values for 'streams'
print(correlation)
```

### **Platform Popularity**

```python
# Finding the highest streamed track
top_track = clean_data .loc[clean_data['streams'].idxmax()]
```

```python
# Plot the data

plt.figure(figsize=(10, 6))
plt.bar(
    ['Spotify Playlists', 'Spotify Charts', 'Apple Playlists'], 
    [top_track['in_spotify_playlists'], 
     top_track['in_spotify_charts'], 
     top_track['in_apple_playlists']], 
    color=['black' , 'gold', 'yellow']
)

# Add labels and title
plt.title(f"Most Popular Track: '{top_track['track_name']}'")
plt.ylabel('Appearances')
plt.xlabel('Platforms')

# Show the plot
plt.show()
```

## **Getting the Key statistics**

```python
filtered_data = clean_data[clean_data['key'] != '"N/A"']
key_stats = filtered_data.groupby('key')['streams'].describe().sort_index()
```

### **Plotting the key statistics**

```python

plt.figure(figsize=(10, 6))

# Plot Mean, Min, and Max values
key_stats[['mean', 'min', 'max']].plot(kind='bar', figsize=(10, 5))

# Adding labels and title
plt.title('Streams Stats per Key (Mean, Min, Max)')
plt.ylabel('Number of Streams')
plt.xlabel('Key')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
